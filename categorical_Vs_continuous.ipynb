### Data Visualization

**Plot the graph for year-wise population**

```python
def formatter(x, pos):
    return str(round(x / 1e9, 1)) + " billion"

fig, ax = plt.subplots(figsize=(12,5))
ax = sns.barplot(x = total_pop_values.index, y = total_pop_values.values, data = total_pop_values, color='royalblue');
ax.yaxis.set_major_formatter(formatter)
ax.set_title('Year-wise Population')
ax.set_xticklabels(labels=years, rotation=90)
plt.grid(axis='x')
```

```python
# Transpose the DataFrame
pop_df = df.T

# Rename the columns to the corresponding years
pop_df.columns = pop_df.iloc[0]
pop_df = pop_df[1:]
```

```python
# Create subplots for each year
years = pop_df.index
years = years.astype(int)
pop_df['years'] = pop_df.index
```

```python
t_pop_df = pop_df.T
t_pop_df = t_pop_df.astype(int)
```

##### Year-wise Top 6 Countries

**Subplots of top-6 from 1960-2022**

```python
def formatter(x, pos):
    return str(round(x / 1e9, 1)) + " billion"

# Create subplots in a 16x4 grid
fig, axes = plt.subplots(16, 4, figsize=(17, 60))

# Iterate through the years
for idx, year in enumerate(years):
    year = str(year)
    sorted_df = t_pop_df.sort_values(by=year, ascending=False)
    sorted_df['Country'] = sorted_df.index
    sorted_df = sorted_df[['Country', year]].head(6)

    # Calculate the row and column for the current subplot
    row = idx // 4
    col = idx % 4

    # Create a bar plot in the current subplot
    ax = axes[row, col]
    sns.barplot(data=sorted_df, x=sorted_df['Country'], y=sorted_df[year], ax=ax)
    ax.yaxis.set_major_formatter(formatter)
    ax.set_title(f'Top 10 Countries by Population in {year}')
    ax.set(xlabel='', ylabel='')
    ax.set_xticklabels(labels=sorted_df.Country, rotation=65)
    
# Clear the last subplot (row 15, column 3)
last_ax = axes[15, 3]
last_ax.cla()  # Clear the content
last_ax.axis('off')  # Turn the axis off

# Adjust layout
plt.tight_layout()
```

```python
roll5_df = total_pop_values.copy()
roll5_df = roll5_df.to_frame()

# Rename the column to Population
roll5_df.rename(columns={0:'Population'}, inplace=True)
```

```python
# Calculate the 5-year rolling percentage change
roll5_df['Rolling_Pct_Change'] = roll5_df['Population'].pct_change(periods=5) * 100

# Plot the data with a blue line
plt.figure(figsize=(13, 4))
plt.plot(roll5_df.index, 
         roll5_df['Rolling_Pct_Change'], 
         label='5-Year Rolling Percentage Change', 
         color='green', 
         marker='o', 
         markersize=3)
plt.ylabel('% Change')
plt.title('5-Year Rolling Percentage Change in Population')
plt.legend()
plt.xticks(rotation=75)
plt.grid(True)
plt.show()
```
